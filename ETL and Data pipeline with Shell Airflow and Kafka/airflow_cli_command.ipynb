Exercise: Apache Airflow CLI
1. Run the command below in the terminal to list all the existing DAGs.
   Answer: airflow dags list
2. Run the command below in the terminal to list all tasks in the DAG named example_bash_operator.
    Answer: airflow tasks list example_bash_operator

3. Run the command below in the terminal to unpause a DAG named tutorial.
    Answer:  airflow dags unpause tutorial

4. Run the command to pause the DAG
    Answer: airflow dags pause tutorial
